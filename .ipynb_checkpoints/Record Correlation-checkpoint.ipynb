# Main Loop for capturing agent information

```python
env = TrackmaniaEnv(observation_space="image",
                        dimension_reduction=6,
                        training_track=None,
                        training_mode="exploration",
                        is_testing=True,
                        render_mode=None,
                        action_mode="human")


obs, _ = env.reset()

step_count = 0
N_steps = 200

asymmetry_list = []
distance_list = []

track_center_position = 592.001

while step_count < N_steps:
    
    action = [0, 0, 0]
    obs, reward, done, _, info = env.step(action)

    time.sleep(0.1)

    _, distance_observation = env.viewer.get_obs()
    asymmetry_list.append(distance_observation)
    distance_list.append(env.position[2])
    

    try:
        if keyboard.is_pressed("q"):
            print("Interrupt")
            break
    except:
        pass

env.close()
```
